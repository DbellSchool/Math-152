```python
# Part 1
# Part A
#c=symbols('c')
#cval=solve(Volc-5,c)


x,y = symbols('x y')
y1 = sin(x)
y2 = cos(x)

# part a --> x axis 

F = integrate(pi*((y2)**2-(y1)**2),(x,0,pi/4)) # bounds are 0 and pi/4
print("the intregral using disks arounnd the x axis is : ",F.evalf())

# Part B disc method --y axis Disk
fs = asin(y)
fc = acos(y)
mid_b = cos(pi/4)
print(mid_b)
print()
F2 = pi*integrate(fs**2,(y,0,mid_b)) 
F3 = pi*integrate(fc**2,(y,mid_b,1)) 

print("the intregral using disks is: ",((F2+F3)).evalf())


#Part C --> y axis Cylidricl shells
F_cyl = 2*pi*integrate((y2-y1)*x,(x,0,pi/4)) 
print("the value using cylinrivlcla shells  y axis is :",F_cyl.evalf())



```

```python
#Part 2
a,b,c,d,x = symbols('a b c d x')

ftop = (a*x**3+b*x**2+c*x+d)
fbot = sqrt(1-x**2)

#Part A 
# take intigral by hand  --> with constants
fegg= (a*x**3+b*x**2+c*x+d)*sqrt(1-x**2)
Vol = pi*integrate(fegg**2,(x,-1,1))
print("the formula for the valume of the egg is:", Vol)


# Part B
ftop_b = ftop.subs({a:-.06,b: .04,c:.1,d:.54})
fegg_sub = fegg.subs({a:-.06,b: .04,c:.1,d:.54})



t = plot(fegg_sub,(x,-1,1),show = False)
t2 = plot(-fegg_sub,(x,-1,1),show = False)
t.extend(t2)
t.show()
# Part C
Vol2 = pi*integrate(fegg_sub**2,(x,-1,1))
print("the formula for the volume of the egg is:", Vol2)
```
