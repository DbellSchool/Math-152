```python
#Part 2
a,b,c,d,x = symbols('a b c d x')

ftop = (a*x**3+b*x**2+c**x)
fbot = (1-x**2)**(1/2)

fegg = (a*x**3+b*x**2+c**x-d*(1-x**2)**(1/2))


ftop_b = ftop.subs({a:-.06,b: .04,c:.1})
fegg_sub = fegg.subs({a:-.06,b: .04,c:.1,d:.54})
a = plot(ftop_b,(x,-1,1))
b = plot(fbot,(x,-1,1))#,show = False)
a.extend(b)
b.show()


b = plot(fegg_sub,(x,-1,1))
```

```python
# Part 3
x,y,r = symbols('x y r')
y1 = (r**2-(x+r/2)**2)**(1/2)
y2 = (r**2-(x-r/2)**2)**(1/2)

y1_sub = y1.subs(r,4)
y2_sub = y2.subs(r,4)

a = plot(y1_sub**2,x)
b = plot(y2_sub**2,x)
a.extend(b)
a.show()


```
